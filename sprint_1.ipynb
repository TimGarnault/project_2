```python
# Let's now group our qualities into 3 categories: low, medium and high quality wine
# add a column 'quality label' with quality score categories
red_wine_df['quality_label']=red_wine_df['quality'].apply(lambda value: 'low' if value <=5 else 'medium' if value <=7 else 'high')

# transforming labels into a categorical data type (specific to pandas) instead of a simple string
red_wine_df['quality_label'] = pd.Categorical(red_wine_df['quality_label'],categories=['low','medium','high'])

red_wine_df
```

```python
# create new column 'quality_label'
white_wine_df['quality_label']=white_wine_df['quality'].apply(lambda value: 'low' if value<=5 else 'medium' if value<=7 else 'high')

# transform the new quality labels into a categorical type (pandas specific) instead of string
white_wine_df['quality_label']=pd.Categorical(white_wine_df['quality_label'], categories=['low','medium','high'])
white_wine_df
```

```python
# merge red wine dataframe with white wine dataframe

#   before merging both dataframes, add a new column with the 'wine type' being either the category 'white' or 'red'
    #   create new column 'wine type'
white_wine_df['wine type']='white'
red_wine_df['wine type']='red'

    #   transform the new quality labels into a categorical type (pandas specific) instead of string
white_wine_df['wine type']=pd.Categorical(white_wine_df['wine type'], categories=['white'])
red_wine_df['wine type']=pd.Categorical(red_wine_df['wine type'], categories=['red'])

ww_df=white_wine_df
rw_df=red_wine_df
    #   add measures to column names:

# ww_df.columns
# Index(['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
    #    'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
    #    'pH', 'sulphates', 'alcohol', 'quality', 'quality_label', 'wine type'],
    #   dtype='object')

# rw_df.columns
# Index(['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
    #    'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
    #    'pH', 'sulphates', 'alcohol', 'quality', 'quality_label', 'wine type'],
    #   dtype='object')


rw_df_2=rw_df.rename(columns={"fixed acidity":"fixed acidity g/dm3","volatile acidity":"volatile acidity g/dm3","citric acid":"citric acid g/dm3","residual sugar":"residual sugar g/dm3","chlorides":"chlorides g/dm3","free sulfur dioxide":"free sulfur dioxide g/dm3","total sulfur dioxide":"total sulfur dioxide g/dm3","density":"density g/cm3","sulphates":"sulphates g/dm3","alcohol":"alcohol % vol"})
ww_df_2=ww_df.rename(columns={"fixed acidity":"fixed acidity g/dm3","volatile acidity":"volatile acidity g/dm3","citric acid":"citric acid g/dm3","residual sugar":"residual sugar g/dm3","chlorides":"chlorides g/dm3","free sulfur dioxide":"free sulfur dioxide g/dm3","total sulfur dioxide":"total sulfur dioxide g/dm3","density":"density g/cm3","sulphates":"sulphates g/dm3","alcohol":"alcohol % vol"})

#     #   add underscore between word
rw_df_2.columns=rw_df_2.columns.str.lower().str.replace(" ","_")
ww_df_2.columns=ww_df_2.columns.str.lower().str.replace(" ","_")

# rw_df_2.info()
# rw_df_2.shape

red_white_wines_df = pd.concat([ww_df_2, rw_df_2], ignore_index=True)

    # red_wine_df.shape
    # # 1599 time 14
    # white_wine_df.shape
    # # 4898 time 14

    # red_white_wines_df.shape
    # # 6497 time 14
red_white_wines_df.info()
```

```python
# Using the red_wine and white_wine datasets, 
# create a table like the one below to compare Descriptive Statistics between red and white wines:

pd.concat([rw_df_2[['fixed_acidity_g/dm3','volatile_acidity_g/dm3','citric_acid_g/dm3','residual_sugar_g/dm3']].describe(), \
           ww_df_2[['fixed_acidity_g/dm3','volatile_acidity_g/dm3','citric_acid_g/dm3','residual_sugar_g/dm3']].describe()], \
            axis=1, keys=["Red Wines Stats","White Wines Stats"])

# comments/observations:
    #   red wine fixed_acidity_g/dm3 > white wine; white wine is less dispersed
    #   red wine volatile_acidity_g (twice) > white wine; white wine is less dispersed
    #   white wine citric_acid_g/dm3 (slightly) > red wine; white wine is less dispersed
    #   white wine residual_sugar_g/dm3 (more than twice) > red wine; white wine is quite dispersed \
    #                                   (any sweet wines included in the white wine data set?)
    
```

```python
pd.concat([rw_df_2[['chlorides_g/dm3','free_sulfur_dioxide_g/dm3','total_sulfur_dioxide_g/dm3','density_g/cm3']].describe(), \
           ww_df_2[['chlorides_g/dm3','free_sulfur_dioxide_g/dm3','total_sulfur_dioxide_g/dm3','density_g/cm3']].describe()], \
            axis=1, keys=["Red Wines Stats","White Wines Stats"])

# comments/observations:
    #   red wine chlorides_g/dm3 (twice) > white wine; white wine is (twice) less dispersed
    #   white wine free_sulfur_dioxide_g/dm3 (more than twice) > red wine; white wine is (almost twice) more dispersed, white wine max is very high
    #   white wine total_sulfur_dioxide_g/dm3 (3 times) > red wine; white wine is more dispersed
    #   density_g/cm3 both white and red wines are similar
```

```python
pd.concat([rw_df_2[['ph','sulphates_g/dm3','alcohol_%_vol']].describe(), \
           ww_df_2[['ph','sulphates_g/dm3','alcohol_%_vol']].describe()], \
            axis=1, keys=["Red Wines Stats","White Wines Stats"])
# comments/observations:
    #   white wine PH > red wine
    #   red wine sulphates_g/dm3 > white wine, red wines values are more dispersed (max value seems quite far from the rest?)
    #   alcohol_%_vol similar for both red and white, white wines values are slightly more dispersed
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# in order to display plot in jupyter.
%matplotlib inline 


# rw_df_2
# ww_df_2
# red_white_wines_df


rw_df_2_number = rw_df_2.select_dtypes(include=('number'))
rw_df_2.dtypes
rw_df_2_number

rw_df_2_corr = rw_df_2_number.corr()

rw_df_2_corr

ax = plt.axes()
# sns.heatmap(rw_df_2_corr.mask(rw_df_2_corr <= 0.3), vmin=-1, cmap='coolwarm', ax=ax, annot=True)
sns.heatmap(rw_df_2_corr, vmin=-1, cmap='coolwarm', ax=ax, annot=True )

ax.set_title('red_wine_corr()')
```

```python
ax = plt.axes()
uppertriangle = np.triu(rw_df_2_corr)
sns.heatmap(rw_df_2_corr, vmin=-1, vmax=1, cmap='coolwarm', ax=ax, annot=True, mask=(rw_df_2_corr >= -0.3) & (rw_df_2_corr <= 0.3) | (uppertriangle))

ax.set_title('Medium_Strong_Positive_Negative_red_wine_corr()')
```

```python
ww_df_2_number = ww_df_2.select_dtypes(include=('number'))
ww_df_2.dtypes
ww_df_2_number

ww_df_2_corr = ww_df_2_number.corr()

ww_df_2_corr


ax = plt.axes()
uppertriangle = np.triu(ww_df_2_corr)
sns.heatmap(ww_df_2_corr, vmin=-1, vmax=1, cmap='coolwarm', ax=ax, annot=True, mask=(ww_df_2_corr >= -0.3) & (ww_df_2_corr <= 0.3) | (uppertriangle))

ax.set_title('Medium_Strong_Positive_Negative_white_wine_corr()')
```
