```python
alcohol_quality = sns.displot(data=rw_df_2, x="alcohol_%_vol", hue="quality_label", kind="kde")

def specs(x, **kwargs):
    ls = {"low":"-","medium":"--","high":"-."}
    plt.axvline(x.median(), linestyle =ls[kwargs.get("label","low")], 
                color = kwargs.get("color", "g"))
    txkw = dict(size=12, color = kwargs.get("color", "g"), rotation=90)
    tx = "median: {:.2f}".format(x.median())
    plt.text(x.median()+0.2, 0.2, tx, **txkw)

alcohol_quality.map(specs,"alcohol_%_vol")

```
